To Classify Movies Based on Ratings

```python
def rating(rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
data['Rating Category']=data['Rating'].apply(rating)
```

# To Count the Number of Comedy Movies in the Dataset

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
data['temp']=list1
```

```python
genre=input("Enter Genre you want to count : ").title()
count=0
for value in data['temp']:
    if genre in value:
        count=count+1
print("Total",genre,"Movies are: ",count)
```
